```python
#Data Science Jobs and Salaries Analysis by Luis Garcilazo
import pandas as pd
df = pd.read_csv('data_science_jobs.csv')
```

```python
#Average salary_in_usd of all jobs for each year
df_per_year = df[['work_year','salary_in_usd']].groupby('work_year').mean().apply(lambda x: round(x,2))
df_per_year
```

```python
#Average Salaries for each year by job title
df_per_year_by_job = df[['work_year','job_title','salary_in_usd']].groupby(['work_year','job_title']).mean().apply(lambda x: round(x,2))
df_per_year_by_job
```

```python
#get frequency of job titles
df_job_title_frequency = df[['job_title']].groupby(['job_title']).size().sort_values(ascending=False)
df_job_title_frequency
```

```python
#10 most common job titles
df_job_title_frequency[0:10]
```

```python
df_salary_in_usd_by_job_title = df[['job_title','salary_in_usd']].groupby('job_title').mean().apply(lambda x: round(x,2))
```

```python
#first 15 job titles with higher salary_in_usd
df_salary_in_usd_by_job_title.sort_values('salary_in_usd',ascending=False)[0:15].apply(lambda x: round(x,2))
```

```python
#first 15 job titles with lowest salary_in_usd
df_salary_in_usd_by_job_title.sort_values('salary_in_usd',ascending=True)[0:15].apply(lambda x: round(x,2))
```

```python
#average salary_in_usd for 10 most common job titles
df_by_frequency = df[['job_title','salary_in_usd']].groupby(['job_title']).agg(['count','mean']).sort_values([('salary_in_usd','count')], ascending=False).apply(lambda x: round(x,2))
df_by_frequency[0:10]
```

```python
#job categories count
df['job_category'].value_counts().plot(kind='bar',title='Count per job category')
```

```python
#average salary_in_usd per category
df_salary_in_usd_by_category = df[['job_category','salary_in_usd']].groupby(['job_category']).mean().apply(lambda x: round(x,2))
df_salary_in_usd_by_category
```

```python
df_salary_in_usd_by_category.plot(kind='bar', title='Avg salary in usd per category')
```

```python
#average salary in usd based on currency
df_salary_in_usd_by_currency = df[['salary_currency','salary_in_usd']].groupby(['salary_currency']).mean().apply(lambda x: round(x,2)).sort_values('salary_in_usd',ascending=False)
df_salary_in_usd_by_currency
```

```python
df_salary_in_usd_by_currency.plot(kind='bar', title='Avg salary in usd per currency')
```

```python
df_salary_by_residence_and_companylocation = df[['employee_residence','company_location','salary_in_usd']].groupby(['employee_residence', 'company_location']).mean().apply(lambda x: round(x,2))
df_salary_by_residence_and_companylocation
```

```python
df_salary_by_residence_and_companylocation.sort_values('salary_in_usd',ascending=False)
```

```python
#avg salary based on experience
df_salary_by_experience = df[['experience_level','salary_in_usd']].groupby(['experience_level']).mean().apply(lambda x: round(x,2)).sort_values('salary_in_usd',ascending=True)
df_salary_by_experience
```

```python
df_salary_by_experience.plot(kind='bar',title='Avg salary in usd based on experience')
```

```python
#Salary based on employement type
df_salary_by_employment_type = df[['employment_type','salary_in_usd']].groupby('employment_type').mean().apply(lambda x: round(x,2)).sort_values('salary_in_usd',ascending=False)
df_salary_by_employment_type
```

```python
df_salary_by_employment_type.plot(kind='bar',title="Avg salary in usd per employment type")
```

```python
#Avg salary based on work setting
df_salary_by_work_setting = df[['work_setting','salary_in_usd']].groupby('work_setting').mean().apply(lambda x: round(x,2)).sort_values('salary_in_usd',ascending=False)
df_salary_by_work_setting
```

```python
df_salary_by_work_setting.plot(kind='bar',title='Avg salary in usd based on work setting')
```

```python
#Avg salary per company size
df_salary_by_company_size = df[['company_size','salary_in_usd']].groupby('company_size').mean().apply(lambda x: round(x,2)).sort_values('salary_in_usd',ascending=False)
df_salary_by_company_size
```

```python
df_salary_by_company_size.plot(kind='bar',title='Avg salary in usd based on company size')
```

```python
#Avg salary in usd based on employment type, work setting, and company size
df_salary_by_type_setting_and_size = df[['employment_type','work_setting','company_size','salary_in_usd']].groupby(['employment_type','work_setting','company_size']).mean().apply(lambda x: round(x,2))
df_salary_by_type_setting_and_size
```

```python
df_salary_by_type_setting_and_size.sort_values('salary_in_usd',ascending=False)
```

```python
df_salary_by_type_setting_and_size.sort_values('salary_in_usd',ascending=False).plot(kind='bar', title='Avg salary in usd based on employment type, work setting, and company size')
```
